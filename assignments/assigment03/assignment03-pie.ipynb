```python
import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Read the CSV file
# Replace 'people-1000.csv' with the correct path to the file
data = pd.read_csv('people-1000.csv')

# Step 2: Extract the email domains
# Assuming that the email column is named 'Email'
# Adjust the column name if necessary
emails = data['Email']

# Function to extract the domain from an email
def extract_domain(email):
    return email.split('@')[-1]

# Create a list of domains
domains = emails.apply(extract_domain)

# Step 3: Count the domains
domain_counts = domains.value_counts()

# Display the top 5 most common domains
top_domains = domain_counts.head(5)

# Step 4: Create a pie chart
plt.figure(figsize=(8, 8))
plt.pie(top_domains, labels=top_domains.index, autopct='%1.1f%%', startangle=90)
plt.title('Distribution of the Top 5 Email Domains')
plt.axis('equal')  # Ensures the pie chart is circular
plt.show()
```
