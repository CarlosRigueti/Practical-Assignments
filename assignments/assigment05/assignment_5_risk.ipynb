```python
import random
import matplotlib.pyplot as plt

# Function to roll 6-sided dice
def roll_dice(num_dice):
    dice_rolls = []
    for _ in range(num_dice):
        dice_rolls.append(random.randint(1, 6))  
    dice_rolls.sort(reverse=True) 
    return dice_rolls

# Function to simulate one battle round between 3 attackers and 2 defenders
def simulate_battle_round():
    # Attacker rolls 3 dice, defender rolls 2 dice
    attacker_dice = roll_dice(3)
    defender_dice = roll_dice(2)
    
    # Initialize the count of troops lost by each side
    attacker_losses = 0
    defender_losses = 0

    # Compare the highest dice rolls
    for i in range(2):  
        if attacker_dice[i] > defender_dice[i]:
            defender_losses += 1  # Defender loses 1 troop
        else:
            attacker_losses += 1  # Attacker loses 1 troop

    return attacker_losses, defender_losses

# Function to simulate multiple battle rounds and collect results
def simulate_multiple_battles(n_rounds):
    attacker_losses_list = []  # List to record attacker losses in each round
    defender_losses_list = []  

    for _ in range(n_rounds):
        attacker_losses, defender_losses = simulate_battle_round()
        attacker_losses_list.append(attacker_losses)
        defender_losses_list.append(defender_losses)

    return attacker_losses_list, defender_losses_list

# Set the number of battle rounds to simulate
n_rounds = 1000
attacker_losses, defender_losses = simulate_multiple_battles(n_rounds)

# Plot the results using a histogram
plt.hist(attacker_losses, bins=range(4), alpha=0.5, color='red', label='Attacker Losses', align='left')
plt.hist(defender_losses, bins=range(3), alpha=0.5, color='blue', label='Defender Losses', align='left')
plt.xlabel('Losses per Battle Round')
plt.ylabel('Frequency')
plt.title('Battle Outcomes in Risk (3 Attackers vs 2 Defenders)')
plt.legend()
plt.show()
```
