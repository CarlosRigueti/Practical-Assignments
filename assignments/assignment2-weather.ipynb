```python
# Week 2 Task: Visualizing Temperature Data

# Developed by: Carlos Rigueti

# Step 1: Import the packages
import pandas as pd
import matplotlib.pyplot as plt
```

```python

# Step 2: Read the CSV file containing weather information
weather_data = pd.read_csv('weatherreadings1.csv')
print (weather_data)
```

```python
# Informations of the dataset.
weather_data.info()
```

```python

# Step 3: Parse the 'reportStartDateTime' column as datetime objects for accurate plotting
weather_data['reportStartDateTime'] = pd.to_datetime(weather_data['reportStartDateTime'])
# Step 4: Generate a temperature plot
plt.figure(figsize=(10, 5))
plt.plot(weather_data['reportStartDateTime'], weather_data['dryBulbTemperature_Celsius'], 'b-', marker='o')
# Step 5: Add descriptive labels and title to the graph
plt.title('Daily Temperature Variation')
plt.xlabel('Date and Time')
plt.ylabel('Temperature (C)')
# Step 6: Display the plot with adjusted formatting
plt.xticks(rotation=45)
plt.grid(visible=True, linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()

```
